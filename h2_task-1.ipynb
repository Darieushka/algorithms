```python
import numpy as np


def m(x,y):
    down = np.array([[np.nan] * (len(y)+1)] * (len(x)+1))
    right = np.array([[np.nan] * (len(y)+1)] * (len(x)+1))
    broadway = np.array([[np.nan] * (len(y)+1)] * (len(x)+1))
    for i in range(0, len(x)+1):
        down[i,0] =- i*sigma
        broadway[i,0] =- i*sigma
    for j in range(0, len(y)+1):
        right[0,j] =- j*sigma
        broadway[0,j] =- j*sigma
    return down, right, broadway, len(x), len(y)

def f(i, j, down, right, broadway):
    s = ma
    if x[i-1] != y[j-1]:
        s = mu
    if str(down[i-1, j])=='nan':
        down[i, j] = broadway[i-1, j] - (ro + sigma)
    else:
        down[i, j]=max(down[i-1, j] - sigma, broadway[i-1, j] - (ro + sigma))
    if str(right[i, j-1]) == 'nan':
        right[i, j] = broadway[i, j-1] - (ro + sigma)
    else:
        right[i, j] = max(right[i, j-1] - sigma, broadway[i, j-1] - (ro + sigma))
    broadway[i,j] = max(down[i, j], right[i, j], broadway[i-1, j-1] + s)
    return down, right, broadway

def m2(down, right, broadway, x_len, y_len):
    for i in range(1, x_len+1):
        for j in range(1, y_len+1):
            down, right, broadway = f(i, j, down, right, broadway)
    return down, right, broadway    
    
def f2(i, j, xx, yy, ch, down, right, broadway):
    if ch == 'b':
        s = ma
        if x[i-1] != y[j-1]:
            s = mu
        if broadway[i-1, j-1] + s == broadway[i, j]:
            return i-1, j-1, x[i-1] + xx, y[j-1] + yy, 'b'
        if broadway[i, j] == down[i, j]:
            return i-1, j-1, x[i-1] + xx, y[j-1] + yy, 'd'
        if broadway[i, j] == right[i, j]:
            return i-1, j-1, x[i-1] + xx, y[j-1] + yy, 'r'
    if ch == 'd':
        if down[i, j] == broadway[i-1, j] - (ro + sigma):
            return i-1, j, x[i-1] + xx, '-' + yy, 'b'
        else:
            return i-1, j, x[i-1] + xx, '-' + yy, 'd'
    if ch == 'r':
        if down[i, j] == broadway[i-1, j] - (ro + sigma):
            return i, j-1, '-' + xx, y[j-1] + yy, 'b'
        else:
            return i, j-1, '-' + xx, y[j-1] + yy, 'r'

def af(x, y):
    down, right, broadway, x_len, y_len = m(x, y)
    down, right, broadway = m2(down, right, broadway, x_len, y_len)
    a = max(broadway[x_len, y_len], down[x_len, y_len], right[x_len, y_len]) 
    if broadway[x_len, y_len] == a:
        ch = 'b'
    elif down[x_len, y_len] == a:
        ch = 'd'
    else:
        ch = 'r'
    xx = ''
    yy = ''
    i = x_len
    j = y_len
    while i + j > 0:
        i, j, xx, yy, ch = f2(i, j, xx, yy, ch, down, right, broadway)
    print(xx)
    print(yy)
    print()
    return None


ma = 1
mu = -1
x = 'TCCCAGTTATGTCAGGGGACACGAGCATGCAGAGAC' 
y = 'AATTGCCGCCGTCGTTTTCAGCAGTTATGTCAGATC'

#test1
ro = 0
sigma = 1
af(x, y)

#test2
ro = 100
sigma = 0.01
af(x, y)

#test3
ro = -0.5
sigma = 0.3
af(x, y)
```
